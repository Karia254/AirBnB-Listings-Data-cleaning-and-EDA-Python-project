**Data Cleaning**

```python
df.isnull().sum()
#Dropping all missing values rows
df.dropna(inplace=True)

df.isnull().sum()
```

```python
df.duplicated().sum()
df[df.duplicated()]
```

**Deleting duplicates**

```python
df.drop_duplicates(inplace=True)
df[df.duplicated()]
```

&nbsp;

**Transforming the data**


```python
df.dtypes
```

```python
df['id']=df['id'].astype(object)
df['host_id']=df['host_id'].astype(object)
df.dtypes
```

&nbsp;

**Explotory Data Analysis**

Univariatte analysis

```python
#Price Distribution
df['price']
```

```python
#Identify outliers in price
sns.boxplot(data=df,x='price')
plt.show()
```

```python
sns.histplot(data=df, x='price')
plt.show()
```

&nbsp;

**Removing Outliers**

```python
#Removing the outliers
data=df[df['price']<2000]
sns.boxplot(data=data, x='price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Price Distribution")
plt.ylabel("Frequency")
sns.histplot(data=data, x='price', bins=100)
plt.show()
```

&nbsp;

```python
#What is the average price in each neighbourhood group
data.groupby(by='neighbourhood_group')['price'].mean()
```

&nbsp;

**Feature Engineering**

```python
#Adding a column that shows pice per bed
data['Price_per_bed']=data['price']/data['beds']
data.head()
```

&nbsp;

```python
#What is the average price per bed in each neighbourhood group
data.groupby(by='neighbourhood_group')['Price_per_bed'].mean()
```

&nbsp;

**Bivariate Analysis**

```python
#What is the average price for the different room types for each neighbourhood group
```

```python
sns.barplot(data=data, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

&nbsp;

What is the ralatiosnhip between price and number of reviews

```python
#Number of reviews and price realationship
plt.figure(figsize=(8,5))
plt.title('Number of reviews depending on price and neighbourhood')
sns.scatterplot(data=data, x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```
